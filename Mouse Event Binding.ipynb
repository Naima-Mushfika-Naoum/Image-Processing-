# Mouse Callback Function

```python
def draw(event,x,y,flags,param):
    
    print("x==",x) 
    print("y==",y)     # x,y -> when we hover mouse,
    print("flags==",flags)  
    # if we perform operation flags 1->left 0->right click
    print("param==",param) # operation is in execution
    
    if event == cv2.EVENT_LBUTTONDBLCLK:         
        # circle will be drawn if left button is double clicked
        cv2.circle(img,(x,y),100,(125,0,200),5)
    if event == cv2.EVENT_RBUTTONDBLCLK:         
        # rectangle will be drawn if left button is double clicked
        cv2.rectangle(img,(x,y),(x+100,y+85),(212,120,255),2)

cv2.namedWindow(winname = "result") # naming windows

img = np.zeros((512,512,3),np.uint8)*255
cv2.setMouseCallback("result",draw) # accept window name and draw function

while True:
    cv2.imshow("result",img)
    if cv2.waitKey(1) & 0xFF == 27:       # 27 esc key
       break 

cv2.destroyAllWindows()
```

## Create a fucntion which help to find cordinate of any pixel and its color

```python
def mouse_event(event, x, y, flg, param):
    print("event==",event)
    print("x==",x)
    print("y==",y)
    print("flg==",flg)
    print("param==",param)
    
    font = cv2.FONT_HERSHEY_PLAIN 
    
    if event == cv2.EVENT_LBUTTONDOWN:        # Single left button click , shows co-ordinate
        print(x,', ' ,y)
           
        cord = ". "+str(x) + ', '+ str(y)
        cv2.putText(img, cord, (x, y), font, 1, (155,125 ,100), 2)
        cv2.imshow('image', img)
        
    if event == cv2.EVENT_RBUTTONDOWN:      # Single right button click, shows color value
        b= img[y, x, 0] #for blue channel is 0
        g = img[y, x, 1] #for green channel is 1
        r = img[y, x, 2] #for red channel is 2
        
        color_bgr = ". "+str(b) + ', '+ str(g)+ ', '+ str(r)
        cv2.putText(img, color_bgr, (x, y), font, 1, (152, 255, 130), 2)
        cv2.imshow('image', img)
        
#img = np.zeros((512, 512, 3), np.uint8) #for black image

img = cv2.imread('image/thor.jpg')
cv2.imshow('image', img)
cv2.setMouseCallback('image', mouse_event)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
