## (e) Perform bit plane slicing on the image in (d).


Bit plane slicing is a method of representing an image with one or more bits of the byte used for each pixel where we divide images in bit plane. 
The three main goals of bit plane slicing is:
<ul>
<li>Converts a gray level image to a binary image.</li>
<li>Represents an image with fewer bits </li>
<li>Reduces the size of the image</li>
<li>Enhances the image by focusing.</li>
</ul>

```python
#Iterate over each pixel 
#change pixel value to binary using np.binary_repr() 
#store it in a list.
I2 =gray2 #gray2 is grayscale image from (d)

h = I2.shape[0] #height of image
w = I2.shape[1] #width of image

lst = []

for i in range(h):
    for j in range(w):
         lst.append(np.binary_repr(I2[i][j] ,width=8)) #no. of bits=8
 
""""We have a list of strings where each string represents binary pixel value. 
To extract bit planes we need to iterate over the strings and 
store the characters corresponding to bit planes into lists.
Multiply with 2^(n-1) and reshape to reconstruct the bit image.
B0(LSB)->0th bit plane & B7(MSB)->7th bit plane """

B0 = (np.array([int(i[7]) for i in lst],
               dtype = np.uint8) * 1).reshape(h,w)
B1 = (np.array([int(i[6]) for i in lst],
               dtype = np.uint8) * 2).reshape(h,w)
B2 = (np.array([int(i[5]) for i in lst],
               dtype = np.uint8) * 4).reshape(h,w)
B3 = (np.array([int(i[4]) for i in lst],
               dtype = np.uint8) * 8).reshape(h,w)
B4 = (np.array([int(i[3]) for i in lst],
                dtype = np.uint8) * 16).reshape(h,w)
B5 = (np.array([int(i[2]) for i in lst],
                dtype = np.uint8) * 32).reshape(h,w)
B6 = (np.array([int(i[1]) for i in lst],
                dtype = np.uint8) * 64).reshape(h,w)
B7 = (np.array([int(i[0]) for i in lst],
                dtype = np.uint8) * 128).reshape(h,w)

# Displaying each bit plane

fig = plt.figure(figsize=(10, 9))
fig.add_subplot(3, 3, 1)
plt.imshow(I2, cmap='gray')
plt.axis('off')
plt.title("Gray_Scale Image")

fig.add_subplot(3, 3, 2)
plt.imshow(B0, cmap='gray')
plt.axis('off')
plt.title("0th plane")

fig.add_subplot(3, 3, 3)
plt.imshow(B1, cmap='gray')
plt.axis('off')
plt.title("1st plane")

fig.add_subplot(3, 3, 4)
plt.imshow(B2, cmap='gray')
plt.axis('off')
plt.title("2nd plane")

fig.add_subplot(3, 3, 5)
plt.imshow(B3, cmap='gray')
plt.axis('off')
plt.title("3rd plane")

fig.add_subplot(3, 3, 6)
plt.imshow(B4, cmap='gray')
plt.axis('off')
plt.title("4th plane")

fig.add_subplot(3, 3, 7)
plt.imshow(B5, cmap='gray')
plt.axis('off')
plt.title("5th plane")

fig.add_subplot(3, 3, 8)
plt.imshow(B6, cmap='gray')
plt.axis('off')
plt.title("6th plane")

fig.add_subplot(3, 3, 9)
plt.imshow(B7, cmap='gray')
plt.axis('off')
plt.title("7th plane")
```

### Analyzing the relative importance of each bit in the image

```python
 
#Concatenate function: cv.hconcat()
finalr = cv.hconcat([B7,B6,B5,B4])
finalv =cv.hconcat([B3,B2,B1,B0])
 
# Vertically concatenate
final = cv.vconcat([finalr,finalv])

plt.imshow(final,cmap='gray')
```

```python
# reading image
I3 = cv.imread("image/3(1).jfif")
img3 = I3[:,:,::-1]


# Creating an array out of pixel values of the image
drop_red_channel = np.array(img3, np.uint8)
drop_green_channel = np.array(img3, np.uint8)
drop_blue_channel = np.array(img3, np.uint8)
  
# Setting the value of every pixel in each channel
drop_red_channel[::, ::, 0] = 0 # dropping red channel
drop_green_channel[::, ::, 1] = 0 # dropping blue channel
drop_blue_channel[::, ::, 2] = 0 # dropping blue channel
  
# Creating an image from the modified array
Red_Channel_zero = Image.fromarray(drop_red_channel)
Green_Channel_zero = Image.fromarray(drop_green_channel)
Blue_Channel_zero = Image.fromarray(drop_blue_channel)

#Displaying Image
fig = plt.figure(figsize=(10, 10))

fig.add_subplot(2, 2, 1)
plt.imshow(img3)
plt.axis('off')
plt.title("Original")

fig.add_subplot(2, 2, 2)
plt.imshow(Red_Channel_zero)
plt.axis('off')
plt.title("Red Channel zero")

fig.add_subplot(2, 2, 3)
plt.imshow(Green_Channel_zero)
plt.axis('off')
plt.title("Green Channel zero")

fig.add_subplot(2, 2, 4)
plt.imshow(Blue_Channel_zero)
plt.axis('off')
plt.title("Blue Channel zero")
```
